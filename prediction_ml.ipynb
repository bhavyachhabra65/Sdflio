```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import plotly
import plotly.express as px
import plotly.graph_objects as go
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
dbd_India=pd.read_csv('https://api.covid19india.org/csv/latest/case_time_series.csv',parse_dates=True)
dbd_India.tail()
```

```python
dbd_India['Recovery Rate(%)']=dbd_India['Daily Recovered']*100/dbd_India['Daily Confirmed']
dbd_India['Death Rate(%)']=dbd_India['Daily Deceased']*100/dbd_India['Daily Confirmed']
dbd_India.tail()
```

```python
fig=go.Figure()
fig.add_trace(go.Scatter(x=dbd_India['Date'],y=dbd_India['Total Confirmed'],mode='lines+markers',name='Total cases'))
fig.update_layout(title_text='Trend of virus cases in India (cummulative cases)',plot_bgcolor='rgb(230,230,230)')
fig.show()

fig=px.bar(dbd_India,x='Date',y='Daily Confirmed',barmode='group',height=400)
fig.update_layout(title_text='Coron virus cases in India on daily basis',plot_bgcolor='rgb(230,230,230)')
fig.show()  %
```

```python
df=pd.read_csv('/Users/bhavyachhabra/Downloads/494724-1196190-bundle-archive/covid_19_data.csv')
df.rename(columns={'ObservationDate':'Date',
          'Country/Region':'Country'},inplace=True)

df_confirmed=pd.read_csv('/Users/bhavyachhabra/Downloads/494724-1196190-bundle-archive/time_series_covid_19_confirmed.csv')
df_recovered=pd.read_csv('/Users/bhavyachhabra/Downloads/494724-1196190-bundle-archive/time_series_covid_19_recovered.csv')
df_deaths=pd.read_csv('/Users/bhavyachhabra/Downloads/494724-1196190-bundle-archive/time_series_covid_19_deaths.csv')

df_confirmed.rename(columns={'Country/Region':'Country'},inplace=True)
df_deaths.rename(columns={'Country/Region':'Country'},inplace=True)
df_recovered.rename(columns={'Country/Region':'Country'},inplace=True)
```

```python
df_deaths.head()
```

```python
df2=df.groupby(['Date','Country','Province/State'])[['Date','Province/State','Country','Confirmed','Deaths','Recovered']]
df2.head()
```

```python
df_india=df.query('Country=="India"').groupby('Date')[['Confirmed','Deaths','Recovered']].sum().reset_index()
df_india['Recovery Rate(%)']=df_india['Recovered']*100/df_india['Confirmed']
df_india['Death Rate(%)']=df_india['Deaths']*100/df_india['Confirmed']
india_confirmed=df_india[['Date','Confirmed']]
india_deaths=df_india[['Date','Deaths']]
india_recovered=df_india[['Date','Recovered']]
india_recoveryRate,india_deathRate=df_india[['Date','Recovery Rate(%)']],df_india[['Date','Death Rate(%)']]
df_india.groupby('Date').sum().tail()
```

```python
confirmed=df_india.groupby('Date').sum()['Confirmed'].reset_index()
recovered=df_india.groupby('Date').sum()['Recovered'].reset_index()
deaths=df_india.groupby('Date').sum()['Deaths'].reset_index()
```

```python
fig=go.Figure()
fig.add_trace(go.Scatter(x=confirmed['Date'],y=confirmed['Confirmed'],mode='lines + markers',name = 'Confirmed',line=dict(color='Blue')))
fig.add_trace(go.Scatter(x=deaths['Date'],y=deaths['Deaths'],mode='lines + markers',name = 'Deaths',line=dict(color='Red')))
fig.add_trace(go.Scatter(x=recovered['Date'],y=recovered['Recovered'],mode='lines + markers',name = 'Recovered',line=dict(color='Green')))
fig.show()
```

```python
from fbprophet import Prophet
```

```python
india_confirmed.columns=['ds','y']
india_confirmed['ds']=pd.to_datetime(india_confirmed['ds'])
```

```python
m=Prophet()
m.fit(india_confirmed)
future=m.make_future_dataframe(periods=64)
future.tail()
```

We can see that model is perdicted till 31st of July

```python
forecast1=m.predict(future)
df_india = pd.concat([df_india, forecast1['yhat']], axis=1)
df_india.rename(columns={'yhat': 'Perdicted Confirmed',},inplace=True)
df_india['Perdicted Confirmed']=df_india['Perdicted Confirmed'].clip(lower=0).astype(int)
forecast1[['ds','yhat','yhat_lower','yhat_upper']].tail(10)
```

```python
india_confirmed=m.plot(forecast1)
```

```python
india_confirmed_forecast_plot=m.plot_components(forecast1)
```
